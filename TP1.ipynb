```python
conn = sqlite3.connect("despesas.db")
cursor = conn.cursor()
```

## Exercício 1

```python
pandas.read_sql_query(" select * from SUBFUNCAO as S where S.COD_SUBFUNCAO in (select COD_SUBFUNCAO from DETALHAMENTO_FUNCAO as D join (select max(COUNT), COD_FUNCAO  from (select count(*) as COUNT, F.COD_FUNCAO  from FUNCAO as F join DETALHAMENTO_FUNCAO as D on F.COD_FUNCAO = D.COD_FUNCAO group by NOME_FUNCAO)) as T on D.COD_FUNCAO = T.COD_FUNCAO)", conn)
```

## Exercício 2

```python
pandas.read_sql_query("select VALOR_DESPESA, NOME_FUNCAO, NOME_SUBFUNCAO from (select * from (select * from (select COD_DETALHAMENTO, VALOR_DESPESA from DESPESA_PUBLICA as DP where DP.VALOR_DESPESA > 200000) as T join DETALHAMENTO_FUNCAO as DF on T.COD_DETALHAMENTO = DF.COD_DETALHAMENTO) as T join FUNCAO as F on T.COD_FUNCAO = F.COD_FUNCAO) as T join SUBFUNCAO as S on S.COD_SUBFUNCAO = T.COD_SUBFUNCAO", conn)
```

## Exercício 3

```python
pandas.read_sql_query("select NOME_ORGAO, VALOR_DESPESA from (select COD_ORGAO, VALOR_DESPESA from DESPESA_PUBLICA where ANO_DESPESA = 2018 order by VALOR_DESPESA desc limit 3) as T join ORGAO as O on T.COD_ORGAO = O.COD_ORGAO", conn)
```

## Exercício 7

```python
pandas.read_sql_query("select NOME_ORGAO, COUNT from (select COD_ORGAO, count(distinct COD_PROGRAMA_GOVERNO) as COUNT from DESPESA_PUBLICA where VALOR_DESPESA > 0 group by COD_ORGAO) as T join ORGAO as O on T.COD_ORGAO = O.COD_ORGAO", conn)
```

## Exercício 9

```python
pandas.read_sql_query("select DP.COD_ORGAO, COD_DESPESA, MES_DESPESA, ANO_DESPESA, MODALIDADE_DESPESA, VALOR_DESPESA from (select COD_ORGAO, QTD_ORGAOS_SUBORDINADOS, max(QTD_ORGAOS_SUBORDINADOS) from (select O.COD_ORGAO, QTD_ORGAOS_SUBORDINADOS from DESPESA_PUBLICA as DP join ORGAO as O on DP.COD_ORGAO = O.COD_ORGAO)) as T join DESPESA_PUBLICA as DP on T.COD_ORGAO = DP.COD_ORGAO", conn)
```

## Exercício 10

```python
pandas.read_sql_query("select DP.COD_DESPESA, DP.MES_DESPESA, DP.ANO_DESPESA, DP.MODALIDADE_DESPESA, VALOR_DESPESA from (select COD_ORGAO, max(VALOR_DESPESA) from DESPESA_PUBLICA where ANO_DESPESA = 2018) as T join DESPESA_PUBLICA as DP on T.COD_ORGAO = DP.COD_ORGAO", conn)
```
